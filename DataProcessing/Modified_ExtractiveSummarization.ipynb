```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Generator_GoodOnly_SmallerQuestion_SummAnswer2 (1).csv")
```

```python
import re

bins = [0, 1000, 1300, 1600, 2000, 3000]
ratios = [0.95, 0.8, 0.65, 0.5, 0.35]

for i in range(len(df)):
    print(i)
    text = df.iloc[i]['Answer']
    text = text.lstrip()

    # copying first sentence
    first = ""
    for char in text:
          if char in ['.', '!', '?']:
            first += char
            break
          else:
            first += char
    #copying last sentence

    last = ""
    punctuation_count = 0
    desired_punctuation_count = 2
    for char in reversed(text):
        if char in ['.', '!', '?',',']:
            punctuation_count += 1
            if punctuation_count == desired_punctuation_count:
                break
        else:
            last = char + last
    #extractive summarization
    length_category = pd.cut([len(text)], bins=bins, labels=False)[0]
    ratio = ratios[length_category]
    raw_answer = model(text, ratio=ratio)


    if first in raw_answer:
      pass
    else:
      #print('first not present\n')
      answer=first + ' ' + raw_answer
    if last in raw_answer:
      pass
    else:
      #print('last not present')
      answer =raw_answer + ' ' + last

    df.at[i, 'SummarizedAnswer'] = answer
    #print(len(df['Answer'][i]))
    #print(len(df['SummarizedAnswer'][i]))
    #print('\n')
```
